```python
import pandas as pd

data=pd.read_csv('train.csv',header=None)
```

```python
data
```

```python
data=data.drop(data.index[0])
```

```python
data
```

```python
data1=data[[0,1,2,5,6,13]]
```

```python
data1
```

```python
target=data[20]
```

```python
x_train=data.iloc[:,:-1]
```

```python
x_train
```

```python
from sklearn.model_selection import train_test_split

x_train1,x_test,y_train,y_test=train_test_split(x_train,target,test_size=0.4,random_state=1)
```

```python
from sklearn import linear_model,metrics

reg=linear_model.LogisticRegression()

reg.fit(x_train1,y_train)
y_pred=reg.predict(x_test)
reg.score(x_test,y_test)
print("Logistic Regression model accuracy(in %):",metrics.accuracy_score(y_test, y_pred)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix


for i in range(1,20):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train1,y_train)
    pred3=knn.predict(x_test)
    print(classification_report(y_test,pred3))
    print('i',i,knn.score(x_test,y_test))
```

```python
y_pred1=knn.predict(x_test)
print("knn model accuracy(in %):",metrics.accuracy_score(y_test, y_pred1)*100)
```

```python
from sklearn.ensemble import RandomForestRegressor 
model = RandomForestRegressor(n_estimators = 200,oob_score = True, n_jobs = -1,random_state =50, max_features = "auto", min_samples_leaf = 50)

model.fit(x_train1,y_train)

model.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()

dtree.fit(x_train1,y_train)

dtree.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(x_train1, y_train)
rfc.score(x_test,y_test)
```
